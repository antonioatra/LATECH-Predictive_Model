```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from lazypredict.Supervised import LazyClassifier
from sklearn.metrics import recall_score, make_scorer, confusion_matrix
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
```

### Puxando os dados

```python
df1 = pd.read_csv('dados/dados_modelo1.csv')
df2 = pd.read_csv('dados/dados_modelo2.csv')
df3 = pd.read_csv('dados/dados_modelo3.csv')
```

### Separando os Data Frames

```python
X1 = df1.drop('Reprovou', axis = 1)
y1 = df1['Reprovou']

X2 = df2.drop('Reprovou', axis = 1)
y2 = df2['Reprovou']

X3 = df3.drop('Reprovou', axis = 1)
y3 = df3['Reprovou']
```

```python
X1['TempoQ3'].fillna(X1['TempoQ3'].mean(), inplace = True)
X2['TempoQ3'].fillna(X2['TempoQ3'].mean(), inplace = True)
X3['TempoQ3'].fillna(X3['TempoQ3'].mean(), inplace = True)
```

### Testando qual modelo melhor se encaixa para cada Data Frame

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1,test_size=0.5, random_state=42)

def recall_1(y_true, y_pred):
    return recall_score(y_true, y_pred, pos_label=1)

clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=recall_1)
models, predictions = clf.fit(X1_train, X1_test, y1_train, y1_test)

print(models)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2,test_size=0.5, random_state=42)

def recall_2(y_true, y_pred):
    return recall_score(y_true, y_pred, pos_label=1)

clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=recall_2)
models, predictions = clf.fit(X2_train, X2_test, y2_train, y2_test)


print(models)
```

```python
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3,test_size=0.5, random_state=42)

def recall_3(y_true, y_pred):
    return recall_score(y_true, y_pred, pos_label=1)

clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=recall_3)
models, predictions = clf.fit(X3_train, X3_test, y3_train, y3_test)


print(models)
```
